## Mean and Std

```python
np.mean(a)
```

```python
mean
```

```python
a
```

$$S^2=\frac{(1-2.5)^2+(2-2.5)^2+(3-2.5)^2+(4-2.5)^2}{4}$$

```python
std = math.sqrt(((1-2.5)**2+(2-2.5)**2+(3-2.5)**2+(4-2.5)**2)/4)
std
```

```python
std = np.sqrt(((a - a.mean())**2/4).sum())
std
```

```python
np.std(a, ddof = 0)
```
