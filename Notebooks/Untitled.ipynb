$$ 2x_1 + 4x_2 + 2x_3 = 16 $$
$$ -2x_1 - 3x_2 + x_3 = -5 $$
$$ 2x_1 + 2x_2 - 3x_3 = -3 $$

$$\begin{bmatrix}
2 & 4 & 2\\
-2 & -3 & 1\\
2 & 2 & -3 
\end{bmatrix}\begin{bmatrix}
x_1\\
x_2\\
x_3\end{bmatrix} = \begin{bmatrix}
16\\
-5\\
-3\end{bmatrix} $$


```python
import numpy.linalg as LA
import numpy as np
```

```python
A = np.array([[2, 4, 2],[-2, -3, 1],[2, 2, -3]])
B = np.array([16, -5, -3])

A
```

```python
sol = LA.solve(A, B)

sol
```

```python
A @ sol
```

```python
np.dot(A, sol)
```

$$\rho = \frac{1}{4}\begin{bmatrix}
2 & 1 & 1\\

$$\rho \vec{v} = \lambda \vec{v}$$

```python
rho = (0.25)*np.array([[2, 1, 1],[1, 1, 0],[1, 0, 1]])

rho
```

```python
D, U = LA.eig(rho)

D
```

```python
U
```

```python
rho @ U[:, 0]
```

```python
D[0] * U[:, 0]
```
